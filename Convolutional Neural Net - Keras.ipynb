```python
import csv
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout,Flatten
from keras.utils import np_utils
import keras.models as kmodel
import keras.layers.convolutional as kconv
import keras.layers.core as klcore
from keras.layers.core import Dense, Activation
from keras.models import Sequential
from keras.optimizers import SGD
from keras.utils import np_utils
import pandas as pd
import numpy as np
```

```python
zip_test = pd.read_csv("zip.test.csv",header=None,sep=' ')
```

```python
zip_test.head()
```

```python
zip_train = pd.read_csv("zip.train.csv",header=None,sep=' ')
```

```python
zip_train.head()
```

```python
#delete last column of NaN
zip_train = zip_train.iloc[:, :-1]
```

```python
train_y = zip_train.ix[:,0].astype("category")
train_x = zip_train.drop(0,axis=1)
```

```python
test_y = zip_test.ix[:,0].astype("category")
test_x = zip_test.drop(0,axis=1)
```

```python
train_x_np = train_x.as_matrix()
train_x_np = train_x_np.astype("float32")
```

```python
test_x_np = test_x.as_matrix()
train_x_np = train_x_np.astype("float32")
```

```python
test_y_np = test_y.as_matrix()
train_y_np = train_y.as_matrix()
```

```python
train_x_np.shape
```

```python
test_x_np.shape
```

```python
train_x_1616=np.reshape(train_x_np,(7291,16,16,1))
```

```python
test_x_1616 = np.reshape(test_x_np,(2007,16,16,1))
```

```python
num_classes = 10
```

```python
train_x_1616 = train_x_1616.astype('float32')
test_x_1616 = test_x_1616.astype('float32')

print('x_train shape:', train_x_1616.shape)
print(train_x_1616.shape[0], 'train samples')
print(test_x_1616.shape[0], 'test samples')

# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(train_y_np, num_classes)
y_test = keras.utils.to_categorical(test_y_np, num_classes)
```

```python
from keras.layers.convolutional import Convolution2D, MaxPooling2D

model = Sequential()
 
model.add(Convolution2D(32, 3, 3, activation='relu', input_shape=(16,16,1)))
model.add(Convolution2D(32, 3, 3, activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
 
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
model.fit(train_x_1616, y_train, validation_data=(test_x_1616, y_test),batch_size=10, epochs=12)
```

```python
score = model.evaluate(test_x_1616, y_test, verbose=0)
print(1 - score[1])
```
